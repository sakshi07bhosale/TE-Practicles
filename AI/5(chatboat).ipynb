```python
def chat_bot_system():
    print("Welcome to MET-Restaurant")
    name = input("Enter Your Name: ")
    print(f"Hello {name}, welcome to MET-Restaurant!\n")
    print(f"What would you like to order, {name}?\n")

    menu_options = ["Rice-Plate", "Samosa", "Vada-Pava", "Chole-Bhature", "Pohe"]
    prices = [50, 25, 25, 55, 25]
    q_count = [0] * len(menu_options)

    while True:
        # Display menu
        for i, item in enumerate(menu_options, start=1):
            print(f"Option {i}: {item}")

        try:
            opt = int(input("\nI would like to have option: "))
        except ValueError:
            print("Please enter a valid number.\n")
            continue

        if opt < 1 or opt > len(menu_options):
            print("Invalid option, please try again.\n")
            continue

        print(f"\nYou confirmed order: {menu_options[opt-1]}")
        q_count[opt-1] += 1

        if q_count[opt-1] >= 5:
            print(f"Limit reached for {menu_options[opt-1]} (max 5).")
            break

        order_more = input("Do you want anything else (yes/no): ").strip().lower()
        print()
        if order_more != "yes":
            break

    # Print order summary
    print("Your Order is:")
    for item, qty in zip(menu_options, q_count):
        if qty > 0:
            print(f"{item} x {qty}")

    # Calculate total bill
    total = sum(qty * price for qty, price in zip(q_count, prices))
    print(f"\nYour total bill is {total}")
    print("\nThanks for your order!")


if __name__ == "__main__":
    chat_bot_system()
```
