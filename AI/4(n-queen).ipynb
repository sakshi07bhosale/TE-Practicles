```python
def print_board(board, N):
    for i in range(N):
        for j in range(N):
            if board[i][j] == 1:
                print("Q", end=" ")
            else:
                print("_", end=" ")
        print()
    print()


def is_safe(board, row, col, N):
    # Check row on left
    for i in range(col):
        if board[row][i] == 1:
            return False

    # Check upper diagonal
    i, j = row, col
    while i >= 0 and j >= 0:
        if board[i][j] == 1:
            return False
        i -= 1
        j -= 1

    # Check lower diagonal
    i, j = row, col
    while i < N and j >= 0:
        if board[i][j] == 1:
            return False
        i += 1
        j -= 1

    return True


def solve_nqueens(board, col, N):
    if col >= N:
        print_board(board, N)
        return True

    res = False
    for i in range(N):
        if is_safe(board, i, col, N):
            board[i][col] = 1
            res = solve_nqueens(board, col + 1, N) or res
            board[i][col] = 0  # backtrack
    return res


if __name__ == "__main__":
    N = int(input("Enter number of Queens: "))
    board = [[0] * N for _ in range(N)]
    if not solve_nqueens(board, 0, N):
        print("No solution exists")
```
