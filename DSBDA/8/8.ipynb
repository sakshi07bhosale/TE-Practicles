```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sb 
```

```python
ds=sb.load_dataset('titanic')
ds.head()
```

```python
sb.displot(x=ds['age'],bins=10,kde=False)
```

```python
sb.jointplot(x=ds['age'],y=ds['fare'],kind='scatter')
sb.jointplot(x=ds['age'],y=ds['fare'],kind='hex')
```

```python
sb.rugplot(ds['fare'])
```

```python
sb.barplot(x=ds['sex'],y=ds['age'])
```

```python
import numpy as np

sb.barplot(x=ds['sex'], y=ds['age'], estimator=np.std)
```

```python
sb.countplot(x=ds['sex'])
```

```python
sb.boxplot(x=ds['sex'],y=ds['age'])
```

```python
sb.boxplot(x=ds['sex'], y=ds['age'], hue=ds['survived'])
```

```python
sb.violinplot(x=ds['sex'],y=ds['age'])
```

```python
sb.violinplot(x=ds['sex'],y=ds['age'],hue=ds['survived'])
```

```python
sb.stripplot(x=ds['sex'],y=ds['age'],jitter=False)
```

```python
sb.stripplot(x=ds['sex'],y=ds['age'],hue=ds['survived'])
```

```python
sb.swarmplot(x=ds['sex'],y=ds['age'])
```

```python
sb.swarmplot(x=ds['sex'],y=ds['age'],hue=ds['survived'])
```
