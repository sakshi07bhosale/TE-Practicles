```python
from sklearn.datasets import fetch_california_housing
```

```python
from sklearn.linear_model import LinearRegression
```

```python
housing = fetch_california_housing(as_frame=True)
```

```python
data = housing.data
```

```python
target=housing.target
```

```python
data.isnull().sum()
```

```python
lm=LinearRegression()
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(data,target,test_size=0.2,random_state=0)
model=lm.fit(xtrain,ytrain)
```

```python
ytrain_pred=lm.predict(xtrain)
ytest_pred=lm.predict(xtest)
```

```python
mse_train=mean_squared_error(ytrain,ytrain_pred)
mse_test=mean_squared_error(ytest,ytest_pred)
```

```python
print("Mean Squred Error for training data:",mse_train)
print("Mean Squred Error for testung data:",mse_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mse=mean_squared_error(ytest,ytest_pred)
print("Mean Squred Error:",mse)
mse=mean_squared_error(ytrain_pred,ytrain)
print("Mean Squred Error:",mse)
```

```python
plt.scatter(ytrain,ytrain_pred,c='blue',marker='o',label='Training data')
plt.scatter(ytest,ytest_pred,c='lightgreen',marker='s',label='Testing data')
plt.xlabel('True Values')
plt.ylabel('Predicted')
plt.title("True value vs Predicted value")
plt.legend(loc='upper left')
plt.show()
```
